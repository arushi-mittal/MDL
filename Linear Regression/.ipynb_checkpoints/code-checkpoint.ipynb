```python
#Import all modules
from math import sqrt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pickle
import random
import matplotlib.pyplot as plt
```

### Loading the train and test data using pickle

```python
with open('./data/test.pkl', 'rb') as f:
    test_data = pickle.load(f)

with open('./data/train.pkl', 'rb') as f:
    train_data = pickle.load(f)
```

### Dividing and Formatting the train and test data

```python
#Divide train set into 10 separate sets

np.random.shuffle(train_data)

data_split = np.split(train_data, 10)
train_x = []
train_y = []
for i in range(10):
    train_x.append(data_split[i][:,0])
    train_x[i] = [[temp] for temp in train_x[i]]
    train_y.append(data_split[i][:, 1])

test_x = test_data[:, 0]
test_x = [[temp] for temp in test_x]
test_y = test_data[:, 1]
```

### Preprocessing the training and test data sets

```python
#training set array [0 - 19] = 10 lists, each list has 800 vectors, each vector is i different coefficients

#test set array [0 - 19] = 80 vectors

train_x_bar = []
test_x_bar = []

#Preprocessing the training set

for i in range (1, 21):
    train_x_bar.append([])
    for j in range (10):
        train_x_bar[i - 1].append(PolynomialFeatures(i).fit_transform(train_x[j]))

#Preprocessing the test set, 0 indexed

for i in range (1, 21):
    test_x_bar.append(PolynomialFeatures(i).fit_transform(test_x))

```

### Building 200 models for 20 polynomial degrees and 10 training sets

```python
#Building models for each degree

#models array = 20 * 10 models (for each degree of each test set there is a model)

#Loop for each degree of the polynomial
    #Loop for each training set
        #LinearRegression.fit(x, y) x = ith degree, jth training set and y = jth training set
models = []
        
for i in range (1, 21):
    models.append([])
    for j in range (10):
        models[i-1].append(LinearRegression().fit(train_x_bar[i - 1][j], train_y[j]))
        
```

### Calculating expectation of f'(x) using predictions made by 200 models on the test set 

```python
#Predictions

#Apply 200 models on test data to get 200 predictions
#total_y

# Loop for degree
# test_x_bar will be constant for a degree => test_x_bar[i]
# y_pred [i - 1th degree] [jth training set] 
# y_pred stores models[i - 1][j].predict(test_x_bar[i - 1])

#average_y_pred stores the average degree-wise for 80 data points averaged across 10 models

y_pred = []
y_total = []
average_y_pred = []
for i in range (1, 21):
    y_total.append(0)
    y_pred.append([])
    for j in range (10):
        y_pred[i - 1].append(models[i - 1][j].predict(test_x_bar[i - 1]))
        y_total[i - 1] += y_pred[i - 1][j]
        
    average_y_pred.append(y_total[i - 1] / 10)
    

y_pred = np.array(y_pred)
```

### Tasks 2: Calculating Bias, Bias squared, Variance, Mean Squared Error and Irreduceable Error

```python
# f_cap_x stores predictions
# f_x stores the actual values 

#bias_2 = bias squared = expectation of the difference between average prediction and actual value squared
#bias  = square root of bias squared
#variance = expectation of difference between predicted value and average of predicted values squared
#mean squared error = average of square of difference in actual and predicted values
#irreduceable error = mean squared error - bias squared - variance

f_cap_x = np.array(average_y_pred)

f_x = np.array(test_y)

xaxis = []
bias_array = []
bias_2_array = []
variance_array = []
temp = []
mean_squared_error_array = []
irreduceable_error_array = []
for i in range (1, 21):
    xaxis.append(i)
    bias_2 = np.mean((f_cap_x[i - 1] - f_x) ** 2)
    bias_2_array.append(bias_2)
    bias = np.mean(np.abs(f_cap_x[i - 1] - f_x))
    bias_array.append(bias)
    

    variance = np.mean((y_pred[i-1] - f_cap_x[i-1].T) ** 2 )
    variance_array.append(variance)
    mean_squared_error = np.mean((y_pred[i-1] - f_x) ** 2 )
    mean_squared_error_array.append(mean_squared_error)
    irreduceable_error = mean_squared_error - bias_2 - variance
    irreduceable_error_array.append(irreduceable_error)
#y_pred over every degree => 10 * 80
```

### Task 3: Tabulating Irreduceable Error

```python
ie_table = pd.DataFrame(np.array([irreduceable_error_array]).T)
ie_table.index = range(1, 21)
ie_table.columns = ["Irreduceable Error"]
ie_table.round(13)
```

### Task 4: Plotting graphs for Bias squared and Variance

```python

plt.plot(xaxis, bias_2_array, label = "bias squared") # Graph of bias squared
plt.plot(xaxis, mean_squared_error_array, label = "mean squared error") # Graph of mean squared error
plt.plot(xaxis, variance_array, label = "variance") # Graph of variance
plt.xticks(np.arange(1, 21, step=2))
plt.legend()
```

### Task: Tabulating the values of bias and variance

```python
bias_variance_table = pd.DataFrame(np.array([bias_array, variance_array]).T)
bias_variance_table.index = range(1, 21)
bias_variance_table.columns = ["Bias", "Variance"]
bias_variance_table.round(2)
```

### Tabulating Values of Mean Squared Error

```python
mse_table = pd.DataFrame(np.array([mean_squared_error_array]).T)
mse_table.index = range(1, 21)
mse_table.columns = ["Mean Squared Error"]
mse_table.round(2)
```

### Plotting graph of irreduceable error

```python
plt.plot(xaxis, irreduceable_error_array, label = "irreduceable error")
plt.show()
```

### Verifying the model by plotting predictions and actual data for each degree

```python
for i in range(1, 21):
    plt.title("Degree = " + str(i))
    plt.plot(test_x, f_cap_x[i - 1], ".")
    plt.plot(test_x, f_x, ".")
    plt.show()
```
